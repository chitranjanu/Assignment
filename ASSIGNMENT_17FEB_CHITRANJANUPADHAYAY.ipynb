```python
# Answer1.
MongoDB is a popular open-source document-oriented NoSQL database that provides a flexible and scalable way to store and retrieve data. Unlike traditional SQL databases that use tables to store data, MongoDB stores data in collections of JSON-like documents with dynamic schemas, making it easy to work with semi-structured and unstructured data.

Non-relational databases, also known as NoSQL databases, are databases that do not use the traditional relational model based on tables, rows, and columns. Instead, they use different data models, such as key-value, document, column-family, or graph models, to store and query data. Non-relational databases are often used for big data, real-time data processing, and web applications that require scalability and high availability.

MongoDB is preferred over SQL databases in scenarios where:

Flexibility: If you have data that has dynamic schemas or is semi-structured, MongoDB provides a flexible way to store and query that data.

Scalability: If you have a large amount of data or need to scale your database quickly and easily, MongoDB can scale horizontally across multiple servers.

Performance: If you need to retrieve data quickly, MongoDB can be faster than SQL databases because it stores data in a format that is optimized for fast reads.

Cloud-native: If you are building cloud-native applications that require high availability, MongoDB's distributed architecture and built-in replication and sharding capabilities make it a good choice.

Overall, MongoDB is a popular choice for modern web and mobile applications that require flexible and scalable data storage and retrieval capabilities.
```

```python
# Answer2.
MongoDB is a document-oriented NoSQL database that provides a range of features designed to make it easier to work with and manage data. Some of the key features of MongoDB include:

Document model: MongoDB stores data in documents that are similar to JSON objects, making it easy to store and work with complex data structures.

Dynamic schema: MongoDB does not enforce a rigid schema like traditional SQL databases, allowing for greater flexibility and easier schema evolution over time.

Scalability: MongoDB can scale horizontally across multiple servers, allowing you to handle large amounts of data and high traffic loads.

Indexing: MongoDB supports multiple indexing types, including unique, compound, and text indexes, allowing for fast and efficient querying of data.

Aggregation framework: MongoDB provides a powerful aggregation framework for performing complex data analysis and manipulation operations, including grouping, sorting, and aggregating data.

MapReduce: MongoDB supports MapReduce, a powerful data processing framework that can be used for a range of tasks, including data analysis, ETL (extract, transform, load), and more.

Replication and sharding: MongoDB provides built-in replication and sharding capabilities for high availability and scalability, allowing you to distribute data across multiple nodes and maintain data consistency.

Ad-hoc queries: MongoDB allows for ad-hoc queries and dynamic queries, making it easy to query data based on specific criteria or conditions.

Transactions: MongoDB provides multi-document transactions that allow for atomic, consistent, isolated, and durable (ACID) operations across multiple documents.
```

```python
# Answer4.
import pymongo

# Connect to MongoDB server
client = pymongo.MongoClient("mongodb://localhost:27017/")

# Create a new database
mydb = client["mydatabase"]

# Create a new collection
mycol = mydb["customers"]

# Insert one document into the collection
mydict1 = { "name": "John", "address": "Highway 37" }
x = mycol.insert_one(mydict1)

# Print the ID of the inserted document
print(x.inserted_id)

# Insert multiple documents into the collection
mylist = [
  { "name": "Amy", "address": "Apple st 652" },
  { "name": "Hannah", "address": "Mountain 21" },
  { "name": "Michael", "address": "Valley 345" },
  { "name": "Sandy", "address": "Ocean blvd 2" },
  { "name": "Betty", "address": "Green Grass 1" },
  { "name": "Richard", "address": "Sky st 331" },
  { "name": "Susan", "address": "One way 98" },
  { "name": "Vicky", "address": "Yellow Garden 2" },
  { "name": "Ben", "address": "Park Lane 38" },
  { "name": "William", "address": "Central st 954" },
  { "name": "Chuck", "address": "Main Road 989" },
  { "name": "Viola", "address": "Sideway 1633" }
]
x = mycol.insert_many(mylist)

# Print the IDs of the inserted documents
print(x.inserted_ids)

# Find one document in the collection
x = mycol.find_one()
print(x)

# Find all documents in the collection
x = mycol.find()
for document in x:
  print(document)
```

```python
# Answer6.
The sort() method is used to sort the results of a MongoDB query based on one or more fields in 
either ascending or descending order. You can pass a sorting parameter to the sort() method that 
specifies the field(s) to sort by and the sort order.

import pymongo

# Connect to MongoDB server
client = pymongo.MongoClient("mongodb://localhost:27017/")

# Create a new database
mydb = client["mydatabase"]

# Create a new collection
mycol = mydb["customers"]

# Query the collection and sort the results by the "name" field in ascending order
mydocs = mycol.find().sort("name", 1)

# Print the sorted documents
for doc in mydocs:
  print(doc)

# Query the collection and sort the results by the "address" field in descending order
mydocs = mycol.find().sort("address", -1)

# Print the sorted documents
for doc in mydocs:
  print(doc)
```

```python
# Answer7.
In MongoDB, the delete_one() and delete_many() methods are used to remove one or multiple documents from a collection that match a specific query, while the drop() method is used to delete an entire collection.

Here's a brief explanation of when and why you would use each of these methods:

delete_one(): This method is used to remove a single document that matches a specific query from a collection. You would use this method when you need to remove a specific document from the collection based on certain criteria. For example, you might use this method to remove a customer record with a specific ID.

delete_many(): This method is used to remove multiple documents that match a specific query from a collection. You would use this method when you need to remove multiple documents from the collection based on certain criteria. For example, you might use this method to remove all customers who live in a specific state.

drop(): This method is used to delete an entire collection, including all of its documents. You would use this method when you need to completely remove a collection and all of its data from the database. For example, you might use this method when you no longer need a collection or when you need to create a new collection with the same name.
```
