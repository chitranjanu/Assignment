```python
# Answer1. 
''' Sure, here are five commonly used plots in Seaborn along with their uses:

Scatter plot: A scatter plot is used to visualize the relationship between two variables. It displays data points as markers on a two-dimensional plane. It can be used to identify patterns or relationships between variables, such as correlation, clustering, or outliers.

Line plot: A line plot is used to visualize the trend of a variable over time or some other continuous variable. It is useful for showing the evolution of data or how one variable changes in relation to another.

Histogram: A histogram is used to visualize the distribution of a single variable. It shows the frequency of values falling within specified intervals, known as bins. It can be used to identify patterns or anomalies in data, such as outliers, skewness, or multimodality.

Box plot: A box plot is used to visualize the distribution of a variable or group of variables. It shows the median, quartiles, and range of the data, as well as any outliers. It can be used to compare the distributions of different groups or to identify potential outliers.

Heat map: A heat map is used to visualize the correlation between two or more variables. It displays the values of the variables as colors in a two-dimensional grid. It can be used to identify patterns or relationships between variables, such as clusters or trends, and to highlight potential areas of interest or concern'''
```

```python
# Answer2.
import seaborn as sns

# Load the fmri dataset
fmri_data = sns.load_dataset("fmri")

# Plot a line plot with event and region
sns.lineplot(data=fmri_data, x="timepoint", y="signal", hue="event", style="region")
```

```python
# Answer3.
import seaborn as sns

# Load the titanic dataset
titanic_data = sns.load_dataset("titanic")

# Plot two box plots
sns.boxplot(data=titanic_data, x="pclass", y="age")
sns.boxplot(data=titanic_data, x="pclass", y="fare")
```

```python
# Answer4.
import seaborn as sns

# Load the diamonds dataset
diamonds_data = sns.load_dataset("diamonds")

# Plot a histogram with the "cut" column as the hue
sns.histplot(data=diamonds_data, x="price", hue="cut", multiple="stack")
```

```python
# Answer5.
import seaborn as sns

# Load the iris dataset
iris_data = sns.load_dataset("iris")

# Plot a pair plot with the "species" column as the hue
sns.pairplot(data=iris_data, hue="species")
```

```python
# Answer6.
import seaborn as sns

# Load the flights dataset
flights_data = sns.load_dataset("flights")

# Pivot the data to create a matrix
flights_matrix = flights_data.pivot(index="month", columns="year", values="passengers")

# Plot the heatmap
sns.heatmap(data=flights_matrix, cmap="YlGnBu", annot=True, fmt="d", cbar_kws={'label': 'Passengers'})
```
